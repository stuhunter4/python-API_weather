```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
file = "Resources/City_Data.csv"
city_data = pd.read_csv(file)
city_data.head()
```

```python
#look for potential outliers in humidity by calculating iqr and upper/lower bounds
humid_quartiles = city_data["Humidity"].quantile([.25, .5, .75])
humid_lowerq = humid_quartiles[0.25]
humid_upperq = humid_quartiles[0.75]
humid_iqr = humid_upperq - humid_lowerq
humid_lowerb = humid_lowerq - (1.5*humid_iqr)
humid_upperb = humid_upperq + (1.5*humid_iqr)
#make a new dataframe without outliers in humidity
clean_city_df = city_data.loc[(city_data["Humidity"] <= humid_upperb) & (city_data["Humidity"] >= humid_lowerb)]
clean_city_df
```

```python
#Convert "Max Temp" data from Kelvin to Fahrenheit
clean_city_df["Max Temp"] = clean_city_df["Max Temp"] * 1.8 - 459.67
clean_city_df
```

```python
#import date class from the datetime module
from datetime import date
#use date.today() to get current local date, strftime() to represent date as dd/mm/YY
today = date.today()
d_today = today.strftime("%d/%m/%Y")

#Latitude vs Temperature(F) Plot; save plotted image as .png
x_axis1 = clean_city_df["Lat"]
y_axis1 = clean_city_df["Max Temp"]
plt.scatter(x_axis1, y_axis1, marker="o", facecolors="turquoise", edgecolors="black", s=25, alpha=0.6)
plt.title(f"City Latitude vs. Max Temperature ({d_today})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid()
plt.savefig("output_data/Latitude_v_Temperature.png")
plt.show()
```

```python
#Latitude vs Humidity Plot; save plotted image as .png
x_axis2 = clean_city_df["Lat"]
y_axis2 = clean_city_df["Humidity"]
plt.scatter(x_axis2, y_axis2, marker="o", facecolors="turquoise", edgecolors="black", s=25, alpha=0.6)
plt.title(f"City Latitude vs. Humidity ({d_today})")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.grid()
plt.savefig("output_data/Latitude_v_Humidity.png")
plt.show()
```

```python
#Latitude vs Cloudiness Plot; save plotted image as .png
x_axis3 = clean_city_df["Lat"]
y_axis3 = clean_city_df["Cloudiness"]
plt.scatter(x_axis3, y_axis3, marker="o", facecolors="turquoise", edgecolors="black", s=25, alpha=0.6)
plt.title(f"City Latitude vs. Cloudiness ({d_today})")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.grid()
plt.savefig("output_data/Latitude_v_Cloudiness.png")
plt.show()
```

```python
#Latitude vs Wind Speed Plot; save plotted image as .png
x_axis4 = clean_city_df["Lat"]
y_axis4 = clean_city_df["Wind Speed"]
plt.scatter(x_axis4, y_axis4, marker="o", facecolors="turquoise", edgecolors="black", s=25, alpha=0.6)
plt.title(f"City Latitude vs. Wind Speed ({d_today})")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid()
plt.savefig("output_data/Latitude_v_WindSpeed.png")
plt.show()
```

```python
north_city_df = clean_city_df.loc[clean_city_df["Lat"] >= 0]
x_axis5 = north_city_df["Lat"]
y_axis5 = north_city_df["Max Temp"]
(slope1, intercept1, rvalue1, pvalue1, stderr1) = st.linregress(x_axis5, y_axis5)
line_eq1 = x_axis5 * slope1 + intercept1
line_ann1 = "y = " + str(round(slope1,2)) + "x + " + str(round(intercept1,2))
plt.scatter(x_axis5, y_axis5)
plt.plot(x_axis5, line_eq1,"r-")
plt.annotate(line_ann1,(7,30),fontsize=15,color="red")
plt.title("Northern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.savefig("output_data/North_Lat_v_MaxTemp_LinR.png")
print(f"The r-value is : {rvalue1}")
plt.show()
```

```python
south_city_df = clean_city_df.loc[clean_city_df["Lat"] <= 0]
x_axis6 = south_city_df["Lat"]
y_axis6 = south_city_df["Max Temp"]
(slope2, intercept2, rvalue2, pvalue2, stderr2) = st.linregress(x_axis6, y_axis6)
line_eq2 = x_axis6 * slope2 + intercept2
line_ann2 = "y = " + str(round(slope2,2)) + "x + " + str(round(intercept2,2))
plt.scatter(x_axis6, y_axis6)
plt.plot(x_axis6, line_eq2,"r-")
plt.annotate(line_ann2,(-40,85),fontsize=15,color="red")
plt.title("Southern Hemisphere - Max Temp vs. Latitude Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.savefig("output_data/South_Lat_v_MaxTemp_LinR.png")
print(f"The r-value is : {rvalue2}")
plt.show()
```
